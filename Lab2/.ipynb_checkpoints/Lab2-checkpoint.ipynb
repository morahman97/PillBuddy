```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
df = pd.read_csv("diseases.csv")
df.head()
```

```python
df.describe()
```

```python
df = df.drop(['geoCode'], axis=1)
```

### Which city makes the most searches regarding diseases?

```python
col = df.sum(axis=1, numeric_only=True)
```

```python
df2 = pd.DataFrame({'Total Searches': col})
```

```python
df = df.join(df2)
```

```python
df.head()
```

```python
df.iloc[df['Total Searches'].idxmax()][0]
```

### Conversely, which city had the fewest searches regarding diseases?

```python
df.iloc[df['Total Searches'].idxmin()][0]
```

### Which year were the most searches done?

```python
year_to_disease = dict()
temp = 0

for col in list(df):
    if '20' in col:
        year = col[0:4]
        if year not in year_to_disease.keys(): 
            year_to_disease[year] = df[col].sum()
        elif year in year_to_disease.keys():
            year_to_disease[year] += df[col].sum()
            
max(year_to_disease, key=year_to_disease.get)
```

### Which disease is searched the most?

```python
disease_to_search = dict()
temp = 0

for col in list(df):
    if len(col) > 3 and 'Total' not in col:
        disease = col[5:]
        if disease not in disease_to_search.keys():
            disease_to_search[disease] = df[col].sum()
        elif disease in disease_to_search.keys():
            disease_to_search[disease] += df[col].sum()
            
max(disease_to_search, key=disease_to_search.get)
```

### Conversely, which disease is searched the least?

```python
min(disease_to_search, key=disease_to_search.get)
```
